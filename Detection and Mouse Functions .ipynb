```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
import pyautogui
import screen_brightness_control as sbc
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path=r'...\yolov5\runs\train\exp\weights\best.pt', force_reload=True)
```

```python
cap = cv2.VideoCapture(0)    
while cap.isOpened():
    ret, frame = cap.read()
    frame = cv2.flip(frame,1)
    results = model(frame)
    
#     print(results.pandas().xywh[0])
    
    xcenter = results.pandas().xywh[0]["xcenter"].tolist()
    ycenter = results.pandas().xywh[0]["ycenter"].tolist()
    
    

    if 0 in results.pandas().xyxy[0]['class'].tolist():
        for x, y in zip(xcenter, ycenter):
            pyautogui.moveTo((x*2.134), (y*1.6))
    elif 1 in results.pandas().xyxy[0]['class'].tolist(): #decrease brightness
        sbc.set_brightness('-10') 
    elif 2 in results.pandas().xyxy[0]['class'].tolist(): #decrease volume
        pyautogui.press('volumedown') 
    elif 3 in results.pandas().xyxy[0]['class'].tolist(): #double click
        pyautogui.doubleClick()
    elif 4 in results.pandas().xyxy[0]['class'].tolist(): #increase brightness
        sbc.set_brightness('+10')
    elif 5 in results.pandas().xyxy[0]['class'].tolist(): #increase volume
        pyautogui.press('volumeup') 
    elif 6 in results.pandas().xyxy[0]['class'].tolist(): #left click
        pyautogui.click()
    elif 7 in results.pandas().xyxy[0]['class'].tolist(): #right click
        pyautogui.click(button='right')
    elif 8 in results.pandas().xyxy[0]['class'].tolist(): #scroll down
        pyautogui.scroll(-50)
    elif 9 in results.pandas().xyxy[0]['class'].tolist(): #scroll up
        pyautogui.scroll(50)
    else:
        print()
        
    cv2.imshow('YOLO', np.squeeze(results.render()))

    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
